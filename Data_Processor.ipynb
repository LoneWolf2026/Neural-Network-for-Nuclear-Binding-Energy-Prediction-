```python
file_path = '/content/drive/MyDrive/AME_Dataset_2021.txt'
content = {}
i = 0

with open(file_path, 'r') as file:
  for line in file:
    line = line.replace("#","")
    line = line.strip()
    line = line.split()
    content[f'Row{i}'] = line
    i += 1
```

```python
import pandas as pd

data = {}
df = pd.DataFrame()

#list_length = []
for b in range(36,len(content)):
  #list_length.append(len(content[f'Row{b}']))

  data[f'Row{b}'] = content[f'Row{b}']
  #if len(data[f'Row{b}']) == 17:
   # length_17 = data[f'Row{b}'].copy()
 # elif len(data[f'Row{b}']) == 16:
   # length_16 = data[f'Row{b}'].copy()
  #elif len(data[f'Row{b}']) == 15:
  #  length_15 = data[f'Row{b}'].copy()
  #elif len(data[f'Row{b}']) == 14:
  #  length_14 = data[f'Row{b}'].copy()
  #print(data[f'Row{b}'])


  B_index = data[f'Row{b}'].index('B-')
  del data[f'Row{b}'][B_index:-3] #delete elements up to 3rd from right
  #print(data[f'Row{b}'])

  # make a step to find the first string from the left of a list

  try:
    data[f'Row{b}'][4] = int(data[f'Row{b}'][4]) #check if 5th element can be converted to integer or not. If yes, then 1st element in list is a unnecessary value
    del data[f'Row{b}'][0]
    data[f'Row{b}'][3] = str(data[f'Row{b}'][3])
    #print(data[f'Row{b}'])
  except:
    #print(data[f'Row{b}'])
    pass

#now all lists are either 13 or 12 elements long

  if len(data[f'Row{b}']) == 13:
    del data[f'Row{b}'][5]
    #print(data[f'Row{b}'])

  elif len(data[f'Row{b}']) == 12:
    #print(data[f'Row{b}'])
    pass

  del data[f'Row{b}'][-3] #Remove the 2nd count of protons closer to the right
  #print(data[f'Row{b}'])

  data_row = pd.DataFrame(data[f'Row{b}'])
  data_row = data_row.T #Take transpose of list

  df = pd.concat([df, data_row],ignore_index=True)


#print(length_17)
#print(length_16)
#print(length_15)
#print(length_14)

#print(max(list_length))
#print(min(list_length))
#print(set(list_length))

#check if distance between [6]
```
