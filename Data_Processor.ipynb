```python
def data_processing(file_path):
  content = {}
  i = 0

  with open(file_path, 'r') as file:
    for line in file:
      line = line.replace("#","")
      line = line.strip()
      line = line.split()
      content[f'Row{i}'] = line

      i += 1

  data = {}

  for b in range(len(content)):
    data[f'Row{b}'] = content[f'Row{b}']

    if 'B-' not in data[f'Row{b}']:
      del content[f'Row{b}'] #'B-' is found in every single row of actual data
    else:
      B_index = content[f'Row{b}'].index('B-')
      del content[f'Row{b}'][B_index:-3] #delete elements up to 4th from right

    print(data[f'Row{b}'])
  #return df
```
