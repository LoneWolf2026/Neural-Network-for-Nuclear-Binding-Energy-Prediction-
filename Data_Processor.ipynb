```python
import pandas as pd

file_path2016 = '/content/drive/MyDrive/AME_Dataset_2016.txt'
file_path2020 = '/content/drive/MyDrive/AME_Dataset_2020.txt'
```

```python
def data_processing(file_path):
  content = {}
  i = 0
  df = pd.DataFrame()

  with open(file_path, 'r') as file:
    for line in file:
      line = line.replace("#","")
      line = line.strip()
      line = line.split()
      content[f'Row{i}'] = line

      i += 1

  data = {}

  for b in range(len(content)):
    data[f'Row{b}'] = content[f'Row{b}']

    if 'B-' not in data[f'Row{b}']:
      del data[f'Row{b}'] #'B-' is found in every single row of actual data
    else:
      B_index = content[f'Row{b}'].index('B-')
      del data[f'Row{b}'][B_index:-3] #delete elements up to 4th from right

      try:
        data[f'Row{b}'][4] = int(data[f'Row{b}'][4]) #check if 5th element can be converted to integer or not. If yes, then 1st element in list is likely an unnecessary value
        del data[f'Row{b}'][0]
        data[f'Row{b}'][3] = str(data[f'Row{b}'][3])
      except:
        pass

    #now all lists are either 13 or 12 elements long

      if len(data[f'Row{b}']) == 13:
        del data[f'Row{b}'][5] #Removes last letters in each row of data

      elif len(data[f'Row{b}']) == 12:
        pass

      del data[f'Row{b}'][-3] #Remove the 2nd count of protons closer to the right

      data_row = pd.DataFrame(data[f'Row{b}'])
      data_row = data_row.T #Take transpose of list
      df = pd.concat([df, data_row],ignore_index=True)

  return df
```

```python
print(data_processing(file_path2016))
print(data_processing(file_path2020))
```

```python
df_2016 = data_processing(file_path2016)
df_2020 = data_processing(file_path2020)

df_2016.to_csv(f'/content/drive/MyDrive/AME_Dataset_2016.csv',index=False)
df_2020.to_csv(f'/content/drive/MyDrive/AME_Dataset_2020.csv',index=False)
```
